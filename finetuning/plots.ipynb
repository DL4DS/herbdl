```python
import matplotlib.pyplot as plt
import json

trainer_file = open("../output/retraining/checkpoint-314916/trainer_state.json")
# trainer_file = open("../output/debug/checkpoint-107000/trainer_state.json")

json_contents = json.load(trainer_file)

trainer_file.close()

steps = [] 
losses = []
logs = json_contents['log_history']

# print(logs)
# for i in logs[:-1]:
#     steps.append(i['step'])
#     losses.append(i['loss'])

for i in logs:
    if i['step']%10 !=0:
        continue
    steps.append(i['step'])
    losses.append(i['loss'])
```

```python
# Create a line plot
plt.figure(figsize=(20, 6))
plt.plot(steps, losses, marker='o', linestyle='-', color='b', label='Training Loss')

# Adding title and labels
plt.title('Training Loss vs. Step Number')
plt.xlabel('Step Number')
plt.ylabel('Loss')
plt.legend()

# Adding grid
plt.grid(True)

# Show plot
plt.show()
```

### Learning rate plot to assess sudden jump

```python
import matplotlib.pyplot as plt
import json

trainer_file = open("../output/retraining/checkpoint-314916/trainer_state.json")
# trainer_file = open("../output/debug/checkpoint-107000/trainer_state.json")

json_contents = json.load(trainer_file)

trainer_file.close()

steps = [] 
lr = []
logs = json_contents['log_history']

for i in logs:
    if i['step']%10 !=0:
        continue
    steps.append(i['step'])
    lr.append(i['learning_rate'])
```

```python
# Create a line plot
plt.figure(figsize=(20, 6))
plt.plot(steps, lr, marker='o', linestyle='-', color='b', label='Training LR')

# Adding title and labels
plt.title('Training Learning rate vs. Step Number')
plt.xlabel('Step Number')
plt.ylabel('Learning Rate')
plt.legend()

# Adding grid
plt.grid(True)

# Show plot
plt.show()
```
